# Data Visualization

How is food related to GPA/Weight? 

We can use data visualization to check and see if our data has any issues, and also to get a higher level view of the data

```python
import pandas as pd

food_df = pd.read_csv("food_coded.csv")
pd.options.display.max_columns = None
display(food_df)
```

How can I get a sense of what the data looks like? Make a boxplot


```python
food_df.describe()
```

```python
# clean data
import seaborn as sns


# this doesn't work because the data is bad


# food_df["GPA"].astype('float64', errors='coerce')

# see what data is not able to be converted into a float
food_df["GPA"].value_counts()

# convert to float, coerce errors into NaN

food_df = food_df.apply(pd.to_numeric, errors='coerce')

food_df = food_df.dropna(subset=['GPA'])

food_df

sns.boxplot(y=food_df["GPA"])

# food_df
# food_df["GPA"].astype(dtype = float, errors = 'coerce')
```

How does GPA differ based on different attributes?

```python
# categorical

sns.boxplot(y=food_df["GPA"], x=food_df["Gender"], data=food_df)
```

```python
# categorical
sns.boxplot(y=food_df["GPA"], x=food_df["drink"], data=food_df)
```

THESE ARE NOT NECESSARILY CAUSAL RELATIONSHIPS
If we change someone's drink association from orange juice to soda, that does not necessarily change their GPA


```python
sns.boxplot(y=food_df["GPA"], x=food_df["drink"], data=food_df)
```

```python
sns.boxplot(y=food_df["weight"], x=food_df["Gender"], data=food_df)
```

```python
sns.boxplot(y=food_df["weight"], x=food_df["drink"], data=food_df, orient="v")
```

So far we've compared numerical data with categorical data. i.e. GPA vs Gender, weight vs drink choice. This only works when there are a reasonable number of categories (it does not work well with favorite food because there are so many of them, which also means fewer observations per category. 


How can we compare numerical data? i.e. GPA vs weight, estimated calories vs weight etc. A Scatterplot!

Predictor vs target variable
target variable is what we are attempting to predict. we are trying to predict weight (although predicting gpa would be useful as well)
predictor variable is what we are using to predict (there will be many of these)

```python


sns.scatterplot(food_df["GPA"],food_df["weight"])
```

It looks like there is almost no correlation between GPA and weight. We are not trying to measure anything yet, just get a sense of what the data looks like. What if we separate out by gender?


```python

sns.scatterplot(food_df["GPA"],food_df["weight"], hue=food_df["Gender"])
```

```python
# to show that there are no 0s
food_df["Gender"].value_counts()
```

```python
sns.scatterplot(food_df["GPA"],food_df["weight"], hue=food_df["Gender"], style=food_df["Gender"])
```

It's tough to tell if there are any correlations, let's make two different graphs, based on gender

```python
female_df = food_df[food_df["Gender"] == 1]

sns.scatterplot(female_df["GPA"],female_df["weight"])
```

```python
male_df = food_df[food_df["Gender"] == 2]

sns.scatterplot(male_df["GPA"],male_df["weight"])
```

Doesn't look like much is going on. I could continue to look at my data in different ways to see if anything interesting showed up, but I'll move on to more quantitative stratgies now
